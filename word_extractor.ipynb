```python
import fitz  # PyMuPDF
import requests
from nltk.tokenize import word_tokenize
from google.colab import files

# 📘 Step 1: Upload PDF Sheet
uploaded = files.upload()
pdf_path = next(iter(uploaded))

# 📘 Step 2: Define helper functions
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()
    return text

def load_common_words():

    common_words_str = """
    the be to of and a in that have i it for not on with he as you do at this but his by from your is
    they we say her she or an will my one all would there their what so up out if about who get which go me english short stories way
    """
    return set(common_words_str.lower().split())

def get_difficult_words_in_order(text, common_words):
    words = word_tokenize(text.lower())
    hard_words = []
    seen = set()
    for word in words:
        if word.isalpha() and word not in common_words and word not in seen:
            hard_words.append(word)
            seen.add(word)
    return hard_words

def get_meaning_and_example(word):
    try:
        url = f"https://api.dictionaryapi.dev/api/v2/entries/en/{word}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            definition = data[0]["meanings"][0]["definitions"][0]["definition"]
            example = data[0]["meanings"][0]["definitions"][0].get("example", None)
            return definition, example
    except:
        return None, None

# 📘 Step 3: Main logic
text = extract_text_from_pdf(pdf_path)
common_words = load_common_words()
hard_words = get_difficult_words_in_order(text, common_words)

print("\n📘 Difficult Words with Meanings and Usage Examples (Paragraph Order):\n")
for word in hard_words[:30]:  # You can change the number as needed
    meaning, example = get_meaning_and_example(word)
    print(f"{word.capitalize()}:")
    print(f"  📖 Meaning: {meaning if meaning else '[Meaning not found]'}")
    print(f"  💬 Example: {example if example else '[Example not found]'}\n")
```
